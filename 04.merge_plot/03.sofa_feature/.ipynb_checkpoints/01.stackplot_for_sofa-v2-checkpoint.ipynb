```python
def ranksums():
    from scipy.stats import ranksums
    for v in [[1,2],[1,3],[2,3]]:
        i = v[0]
        j = v[1]
        print(i,j)
        data1 = data_groups[i]
        data2 = data_groups[j]
        differences = [x - y for x, y in zip(data1, data2)]
        statistic, p_value = ranksums(data1, data2)
        print("秩和检验统计量:", statistic)
        print("p值:", p_value)
```
