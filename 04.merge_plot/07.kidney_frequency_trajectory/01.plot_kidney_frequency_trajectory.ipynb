# kidney trajectory

```python
import matplotlib.pyplot as plt
fea_lst = ["urineoutput", "creatinine"]
lab_dict = {"urineoutput": "urineoutput (ml)",  "creatinine": "creatinine (mg/dl)"}
```

```python
for fea in fea_lst:
    df_type = pd.read_csv("/public/hanl/jupyter_dir/kidney_sepsis_penotype_v3/01.MIMICIV_SAKI_trajCluster/df_mixAK_fea4_C3.csv")
    df_type_filt = df_type.loc[:,["stay_id","groupHPD"]]
    df_type_filt = df_type_filt.drop_duplicates()
    df_fea = pd.read_csv("/public/hanl/jupyter_dir/kidney_sepsis_penotype_v3/01.MIMICIV_SAKI_trajCluster/sk_feature_timescale_Fb2.csv")
    choose_time_lst = [-2,-1,1,5,9,13,17,21,25]
    df_fea =  df_fea[df_fea["time"].isin(choose_time_lst)] 
    df = pd.merge(df_type_filt,df_fea, how="inner",on="stay_id")
    lw=2
    fig = plt.figure(figsize=(4,2.5))
    df1 = df[df["groupHPD"]==1][[fea,"time"]].groupby('time').agg({fea:"mean"}).reset_index()
    df2 = df[df["groupHPD"]==2][[fea,"time"]].groupby('time').agg({fea:"mean"}).reset_index()
    df3 = df[df["groupHPD"]==3][[fea,"time"]].groupby('time').agg({fea:"mean"}).reset_index()
    plt.plot(df1.time.tolist(),df1[fea].tolist(),c='#91bfdb',marker='o',lw=lw)
    plt.plot(df2.time.tolist(),df2[fea].tolist(),c='#91cf60',marker='o',lw=lw)
    plt.plot(df3.time.tolist(),df3[fea].tolist(),c='#fc8d59',marker='o',lw=lw)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.title("MIMIC-IV")
    plt.xlabel('Time after SA-AKI onset (6h windows)',fontsize=14)
    plt.ylabel(lab_dict[fea],fontsize=14)
    outfile = "mimic_" + fea + ".pdf"
    plt.savefig(outfile, bbox_inches = 'tight')
    plt.show()
```
