````r
obs_data <- setDT(df_select)
outcome_type <- 'survival'
outcome_name <- 'mortality_28d'  # 'Surv(survival_28day, mortality_28d)' 

id <- 'id'
time_name <- 'time'
time_points <- 8
histories <- c(lagged, lagavg) # lagged, lagavg, cumavg   
histvars <- list(c('treat','sofa'), c('sofa'))
basecovs <- c('weight') 
covnames <- c( 'sofa','treat') # 连续协变量
covtypes <- c( 'normal', 'categorical' )  #"binary", "normal", "categorical", "bounded normal", "zero-inflated normal", "truncatednormal", "absorbing", "categorical time", and "custom".
covparams <- list(covmodels = c(sofa ~ lag1_treat + lag_cumavg1_sofa + weight + time,
                                treat ~ lag1_treat + sofa + lag_cumavg1_sofa + weight + time))
ymodel <- mortality_28d ~ treat + sofa + lag1_treat + lag1_sofa + weight + time

intvars <- list('treat', 'treat', 'treat','treat', 'treat', 'treat', 'treat','treat', 'treat', 'treat','treat', 'treat', 'treat','treat', 'treat', 'treat' )
interventions <- list(list(c(static, rep(1, time_points))), list(c(static, rep(2, time_points))), list(c(static, rep(3, time_points))), 
                      list(c(static, rep(4, time_points))), list(c(static, rep(5, time_points))), list(c(static, rep(6, time_points))), 
                      list(c(static, rep(7, time_points))), list(c(static, rep(8, time_points))), list(c(static, rep(9, time_points))), 
                      list(c(static, rep(10, time_points))), list(c(static, rep(11, time_points))), list(c(static, rep(12, time_points))),
                      list(c(static, rep(13, time_points))), list(c(static, rep(14, time_points))), list(c(static, rep(15, time_points))), list(c(static, rep(16, time_points))))

int_descript <- c('-20000_-3000', '-3000_-2500', '-2500_-2000', '-2000_-1500', '-1500_-1000', '-1000_-500', '-500_-200', '-200_0', '0_200', '200_500', '500_1000', '1000_1500', '1500_2000', '2000_2500', '2500_3000', '3000_20000')
ncores <- 4 # parallel::detectCores()-12
nsimul <- 5000
#nsamples <- 200

gform_res <- gformula(obs_data = obs_data,
                          outcome_type = outcome_type, id = id, time_points=time_points,
                          time_name = time_name, covnames = covnames,
                          outcome_name = outcome_name, covtypes = covtypes,
                          covparams = covparams, ymodel = ymodel,
                          intvars = intvars, interventions = interventions, int_descript = int_descript,
                          histories = histories, histvars = histvars, basecovs = basecovs,
                          seed = 1234, parallel = TRUE, ncores=ncores, nsimul = nsimul) #  nsamples = nsamples,
````

# 说明书：

````r
# The nonparametric estimate (NP risk) of the natural course risk is 0.506 and close to the parametric g-formula estimate of this risk. This supports (but does not guarantee) the absence of gross model misspecification.
系数 bi 称为危险比率（HR，hazard ratio）。bi 值大于零，或相当于风险比率大于1，表明随着第 i 个协变量值的增加，事件风险增加，因此生存时间缩短。

换句话说，风险比大于1表示协变量与事件概率正相关，因此与存活时间负相关。
总之，
HR=1：无影响
HR<1：危害降低
HR>1：危险增加
````

````r
# 文中提到的四个指标分别是：
# 1. NP risk：这是一种表示新发患病率的指标，通常表示为“number needed to treat to prevent one new case”。
# 该指标描述了在治疗组中，为防止一个新发病例需要治疗多少个人的数量。

# 2. g-form risk：这是一种相对危险度的指标，通常表示为“generalized formula for risk ratio”。
# 该指标比较了治疗组和对照组的患病率之间的差异，以确定治疗是否具有显著的治疗效果。

# 3. Risk ratio：这是一种相对危险度的指标，通常表示为“relative risk”。
# 该指标比较了治疗组和对照组的患病率之间的比率，以确定治疗是否具有显著的治疗效果。

# 4. Risk difference：这是一种绝对危险度的指标，通常表示为“absolute risk difference”。
# 该指标比较了治疗组和对照组的患病率之间的实际差异，以确定治疗是否具有显著的治疗效果。
````


红框中的Intervention 0代表自然进程，1和2分别代表参数“interventions”中设置的两种场景，即从未治疗与一直治疗。绿框中的NP risk代表非参数方法计算的结局发生率；g-form risk代表利用parametric g-formula方法参数建模估计出的结局发生率；Risk ratio代表1和2模拟场景与自然进程结局发生率的比值。本次分析结果表示，从未治疗和始终治疗两种干预场景的结局发生率分别为自然进程下的1.45和0.46倍。
