Together with source control and treatment with antimicrobials, the administration of fluids and vasopressors are key management strategies in SA-AKI
- 我们建议每日和累计监测体液平衡(1C级)并发非肾器官功能障碍，为SA-AKI患者的液体管理策略提供信息
- 我们建议液体方案和监测尿量和肾功能的频率考虑AKI的严重程度和进展速度
- 我们建议根据患者酸碱和电解质失衡的需要来选择液体
- 我们建议根据个体患者的生化特征，使用平衡溶液和0.9%生理盐水进行复苏，同时密切监测其生化效果
- Albumin and bicarbonate might be of benefit in SA-AKI\
- 我们建议对液体负荷过重的患者使用利尿剂


我们的目的：探究肾损伤程度与液体复苏量的关系？与死亡的关系？

预期结果：肾损伤程度越重，液体复苏量越轻
