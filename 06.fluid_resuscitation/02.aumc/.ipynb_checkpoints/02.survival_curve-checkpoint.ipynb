```r
library(MatchIt)
library(survival)
library(survminer)
library(tidyr)
library(tableone)
```

```r
#函数
survival_curve <- function(data,title,survival_day,expire_flag){
    df_select <- data
    fit <- survfit(Surv(survival_day,expire_flag) ~ groupHPD, data = df_select)

    ggsurvplot(fit, data = df_select, 
               surv.median.line = "hv",  # 增加中位生存时间
               conf.int = TRUE, # 增加置信区间
              # risk.table = TRUE, # 绘制累计风险曲线
#               title=title,
               pval = TRUE,
               #add.all = TRUE, # 添加总患者生存曲线
               xlab = "Follow up time(day)",
               legend = c(0.2,0.2)) # 指定图例位置
  #  ggsave(paste0("./plot_survival/plot/",title,".pdf"), print(pp))
}
```

# surv_cutpoint

```r
df_survival <- read.csv("aumc_fluid_survival.csv",as.is = TRUE)
df_survival <- as.data.frame(df_survival)
df_select =  df_survival[,c("survival_28day","mortality_28d","colloid_bolus_cum", "groupHPD","colloid_bolus_cum_kg")]
library(ggplot2)
p <- ggplot(df_select, aes(x = colloid_bolus_cum_kg)) + ggtitle("AUMCdb")
p + geom_density(color = "black", fill = "gray")
```

```r
density_obj <- density(df_select$colloid_bolus_cum_kg)  
cdf <- cumsum(density_obj$y)  # 计算累积分布函数
x_half <- density_obj$x[which.min(abs(cdf - 0.5))]# 找到累积分布函数等于1/2的横坐标
x_half
```

```r
#  Determine the optimal cutpoint of variables
res1 <- surv_cutpoint(df_select, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"))
print(summary(res1))

df_select_filt  <- subset(df_select, colloid_bolus_cum_kg >= summary(res1)$cutpoint)
res2 <- surv_cutpoint(df_select_filt, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"))
print(summary(res2))

df_select_filt1  <- subset(df_select_filt, colloid_bolus_cum_kg >= summary(res2)$cutpoint)
res3 <- surv_cutpoint(df_select_filt1, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"))
print(summary(res3))

df_select_filt2  <- subset(df_select_filt1, colloid_bolus_cum_kg >= summary(res3)$cutpoint)
res4 <- surv_cutpoint(df_select_filt2, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"))
print(summary(res4))

#####
df_select_filt  <- subset(df_select, colloid_bolus_cum_kg <= summary(res1)$cutpoint)
res2 <- surv_cutpoint(df_select_filt, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"))
print(summary(res2))

df_select_filt1  <- subset(df_select_filt, colloid_bolus_cum_kg <= summary(res2)$cutpoint)
res3 <- surv_cutpoint(df_select_filt1, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"))
print(summary(res3))
```

```r
#  Determine the optimal cutpoint of variables
res1 <- surv_cutpoint(df_select, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"),minprop =0.3)
print(summary(res1))

df_select_filt  <- subset(df_select, colloid_bolus_cum_kg >= summary(res1)$cutpoint)
res2 <- surv_cutpoint(df_select_filt, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"),minprop =0.3)
print(summary(res2))

df_select_filt1  <- subset(df_select_filt, colloid_bolus_cum_kg >= summary(res2)$cutpoint)
res3 <- surv_cutpoint(df_select_filt1, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"),minprop =0.3)
print(summary(res3))

#####
df_select_filt  <- subset(df_select, colloid_bolus_cum_kg <= summary(res1)$cutpoint)
res2 <- surv_cutpoint(df_select_filt, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"),minprop =0.3)
print(summary(res2))

df_select_filt1  <- subset(df_select_filt, colloid_bolus_cum_kg <= summary(res2)$cutpoint)
res3 <- surv_cutpoint(df_select_filt1, time = "survival_28day", event = "mortality_28d",variables = c("colloid_bolus_cum_kg"),minprop =0.3)
print(summary(res3))
```
