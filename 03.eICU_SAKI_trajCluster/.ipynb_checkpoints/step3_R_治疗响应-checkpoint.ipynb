```r
library(survival)
library(survminer)
library(tidyr)
```

# 导入数据

```r
df_survival <- read.csv("df_vaso_stat.csv",as.is = TRUE)
df_survival <- as.data.frame(df_survival)
df_survival[1:2,]
```

# 函数
- 生存图

```r
survival_curve <- function(data,title){
    df_select <- data
    fit <- survfit(Surv(hospital_survival_day,hospital_expire_flag) ~ class, data = df_select)

    ggsurvplot(fit, data = df_select, 
               surv.median.line = "hv",  # 增加中位生存时间
               conf.int = TRUE, # 增加置信区间
             #  risk.table = TRUE, # 绘制累计风险曲线
               title=title,
               pval = TRUE,
               #add.all = TRUE, # 添加总患者生存曲线
               xlab = "Follow up time(day)",
               legend = c(0.2,0.2)) # 指定图例位置
}
```

# 三组生存曲线

```r
df_select =  df_survival[,c("stay_id","hospital_survival_day","hospital_expire_flag","groupHPD")]
df_select = df_select[!duplicated(df_select), ]
df_select["class"] <- df_select["groupHPD"]
survival_curve(df_select,title="The survival curve of three group")
```

# 升压药

```r
df_select =  df_survival[,c("stay_id","hospital_survival_day","hospital_expire_flag","groupHPD")]
df_select = df_select[!duplicated(df_select), ]
df_select["class"] <- df_select["groupHPD"]

fit <- survfit(Surv(hospital_survival_day,hospital_expire_flag) ~ class, data = df_select)

ggsurvplot(fit, data = df_select, 
           surv.median.line = "hv",  # 增加中位生存时间
           conf.int = TRUE, # 增加置信区间
           title="all vaso",
         #  risk.table = TRUE, # 绘制累计风险曲线
           pval = TRUE,
           #add.all = TRUE, # 添加总患者生存曲线
           xlab = "Follow up time(day)",
           legend = c(0.2,0.2)) # 指定图例位置
```

## 使用升压药与否

```r
df_select =  df_survival[,c("stay_id","hospital_survival_day","hospital_expire_flag","with_vaso_instayid","groupHPD")]
df_select = df_select[!duplicated(df_select), ]
df_select = tidyr::unite(df_select, "class", groupHPD, with_vaso_instayid)#, remove = FALSE)
survival_curve(df_select,title="whether use vaso")
```

```r
survival_curve <- function(data,title){
    df_select <- data
    fit <- survfit(Surv(hospital_survival_day,hospital_expire_flag) ~ class, data = df_select)

    ggsurvplot(fit, data = df_select, 
               surv.median.line = "hv",  # 增加中位生存时间
               conf.int = TRUE, # 增加置信区间
               risk.table = TRUE, # 绘制累计风险曲线
               title=title,
               pval = TRUE,
               #add.all = TRUE, # 添加总患者生存曲线
               xlab = "Follow up time(day)",
               legend = c(0.2,0.2)) # 指定图例位置
}
```

```r
df_select =  df_survival_fuild[,c("stay_id","hospital_survival_day","hospital_expire_flag","is_bolus","groupHPD")]
df_select = df_select[!duplicated(df_select), ]
df_select = tidyr::unite(df_select, "class", groupHPD, is_bolus)#, remove = FALSE)
survival_curve(df_select,title="whether have fuild bolus" )
```
