```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_fea = pd.read_csv("./aa.csv")
df_fea = df_fea.loc[:,["stay_id","time"]].drop_duplicates()
```

```python
len(df_fea.stay_id.unique())
```

```python
df_ = df_fea
df_ = pd.DataFrame(df_["time"].value_counts()).reset_index()
df_["stayid_count"] = 7282
df_["percent"] = round(df_["time"]/df_["stayid_count"]*100,2)
```

```python
df = df_[(df_["index"]<30) & (df_["index"]>=-5)]
df = df.sort_values("index")

plt.plot(df["index"], df["percent"],"o-", c="#7570b3")
plt.ylabel("percent(%)")
plt.xlabel("time")
plt.show()
```
