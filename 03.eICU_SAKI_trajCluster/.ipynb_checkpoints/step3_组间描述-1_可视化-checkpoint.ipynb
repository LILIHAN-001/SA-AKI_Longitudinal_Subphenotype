## 颜色定义

```python
# 定义一个函数，将十六进制颜色转换为RGB元组
def hex_to_rgb(hex):
    hex = hex.lstrip('#')
    return tuple(int(hex[i:i+2], 16) for i in (0, 2, 4))

# 定义一个函数，将RGB元组转换为十六进制颜色
def rgb_to_hex(rgb):
    return '#%02x%02x%02x' % rgb

# 定义一个函数，计算两个RGB元组之间的线性插值
def linear_gradient(rgb1, rgb2, n):# 初始化渐变色数组    
    gradient = []
    r_step = (rgb2[0] - rgb1[0]) / (n - 1)# 计算每个通道的步长
    g_step = (rgb2[1] - rgb1[1]) / (n - 1)
    b_step = (rgb2[2] - rgb1[2]) / (n - 1)
    for i in range(n):# 循环生成渐变色
        r = round(rgb1[0] + r_step * i)
        g = round(rgb1[1] + g_step * i)
        b = round(rgb1[2] + b_step * i)
        gradient.append(rgb_to_hex((r, g, b)))
    return gradient

def get_time_color(color_len,class_):
    if class_ == "1":# 绿色
        start_hex = '#f7fcfd' #'' 
        end_hex = '#00441b' #'' 
    if class_ == "2": # 红色
        start_hex = '#fff5f0' 
        end_hex = '#a50f15' 
    if class_ == "3": #蓝色
        start_hex = '#fcfbfd' 
        end_hex = '#3f007d' 
    start_rgb = hex_to_rgb(start_hex) # 转换为RGB元组
    end_rgb = hex_to_rgb(end_hex)
    gradient_array = linear_gradient(start_rgb, end_rgb, color_len) # 调用函数生成渐变色数组
    return gradient_array
color_map_dict = {1:"#00441b",2:"#e34a33",3:"#524C90"}
get_time_color(3,"1")
```

```python
def plot_tsne_3d(X,LABELS,colorbar):
    import numpy as np
    from sklearn.manifold import TSNE
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D
    classes = ["C1","C2","C3"]
    X_tsne_t = TSNE(n_components=3,random_state=50,perplexity=30).fit_transform(X)
    fig = plt.figure(figsize=(8, 8))
    ax= Axes3D(fig)
    scatter = ax.scatter(X_tsne_t[:,0], X_tsne_t[:,1], X_tsne_t[:, 2],c=colorbar,s=10)  # c是颜色，可以新加一列，按时间来渐变; # s即size
    plt.legend(handles=ax.scatter(X_tsne_t[:,0], X_tsne_t[:,1], X_tsne_t[:, 2],c=LABELS,s=10).legend_elements()[0], labels=classes)
    name = "./3D_visualization/" + data_type_ + "_012_" + colorbarname  
    plt.savefig(name)
    plt.show()
    
df = df_fea_base_add
X_scaled = scale(df[fea_select])
y_ = df["groupHPD"].values
y_.astype(int)
plot_tsne_3d(X_scaled, y_,colorbar)
```

## 二维可视化

```python
def plot_tsne(X,LABELS):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    from sklearn.manifold import TSNE
    colors = ['#fc8d59','#91cf60','#524C90','#e34a33','#845868','#e24fff','#ffffbf'] #['#fc8d59','#ffffbf','#91cf60','#524C90','#845868','#e24fff']
    n_cluster = len(list(set(LABELS)))  # 参数定义
    tsne = TSNE(n_components=2, perplexity=30,random_state=50)#50
    Y = tsne.fit_transform(X)
    Y = pd.DataFrame(Y,columns=['x','y'])
    label = pd.DataFrame(LABELS,columns=['label'])
    df_Y_with_label = pd.concat([Y, label],axis=1) # 带有聚类label的数据集生成
    i = 1 
    fig = plt.figure(figsize=(6,4))
    while i <= n_cluster:
        Y_with_onelabel = df_Y_with_label.loc[df_Y_with_label['label']==i].values
        plt.scatter(Y_with_onelabel[:, 0], Y_with_onelabel[:,1], color=colors[i],s=10, alpha=0.8)
        i += 1
    plt.legend(["C1","C2","C3"])
    plt.title("t-SNE")
    plt.show()  # 画图
    
def plot_umap(X,LABELS):
    import umap.umap_ as umap
    colors = ['#fc8d59','#91cf60','#524C90','#e34a33','#845868','#e24fff','#ffffbf']
    n_cluster = len(list(set(LABELS)))  # 参数定义
    
    Y = umap.UMAP( min_dist=0.001 ).fit_transform(X)
    Y = pd.DataFrame(Y,columns=['x','y'])
    label = pd.DataFrame(LABELS,columns=['label'])
    df_Y_with_label = pd.concat([Y, label],axis=1) # 带有聚类label的数据集生成
    
    i = 1 
    fig = plt.figure(figsize=(6, 4))
    while i <= n_cluster:
        Y_with_onelabel = df_Y_with_label.loc[df_Y_with_label['label']==i].values
        plt.scatter(Y_with_onelabel[:, 0], Y_with_onelabel[:,1], color=colors[i],s=10, alpha=1)
        i += 1
    plt.legend(["C1","C2","C3"])
    plt.title('UMAP')
    plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import scale
df = df_fea_base_add
X_scaled = scale(df[fea_select])
y_ = df["groupHPD"].values
y_.astype(int)
plot_tsne(X_scaled, y_)
```
