```python
def plot_tsne(X,LABELS):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    from sklearn.manifold import TSNE
    colors = ['#fc8d59','#91cf60','#524C90','#e34a33','#845868','#e24fff','#ffffbf'] #['#fc8d59','#ffffbf','#91cf60','#524C90','#845868','#e24fff']
    n_cluster = len(list(set(LABELS)))  # 参数定义
     
    tsne = TSNE(random_state=50)
    Y = tsne.fit_transform(X)
    Y = pd.DataFrame(Y,columns=['x','y'])
    label = pd.DataFrame(LABELS,columns=['label'])
    df_Y_with_label = pd.concat([Y, label],axis=1) # 带有聚类label的数据集生成

    i = 1 
    fig = plt.figure(figsize=(12, 8))
    while i <= n_cluster:
        Y_with_onelabel = df_Y_with_label.loc[df_Y_with_label['label']==i].values
        plt.scatter(Y_with_onelabel[:, 0], Y_with_onelabel[:,1], color=colors[i],s=16, alpha=0.8)
        i += 1
    plt.legend(["C1","C2","C3"])
    plt.title("t-SNE")
    plt.show()  # 画图
```

```python
def plot_umap(X,LABELS):
    import umap.umap_ as umap
    colors = ['#fc8d59','#91cf60','#524C90','#e34a33','#845868','#e24fff','#ffffbf']
    n_cluster = len(list(set(LABELS)))  # 参数定义
    
    Y = umap.UMAP( min_dist=0.001 ).fit_transform(X)
    Y = pd.DataFrame(Y,columns=['x','y'])
    label = pd.DataFrame(LABELS,columns=['label'])
    df_Y_with_label = pd.concat([Y, label],axis=1) # 带有聚类label的数据集生成
    
    i = 1 
    fig = plt.figure(figsize=(12, 8))
    while i <= n_cluster:
        Y_with_onelabel = df_Y_with_label.loc[df_Y_with_label['label']==i].values
        
        plt.scatter(Y_with_onelabel[:, 0], Y_with_onelabel[:,1], color=colors[i],s=16, alpha=1)
        i += 1
    plt.legend(["C1","C2","C3"])
    plt.title('UMAP')
    plt.show()
```

```python
# 一张图为一个特征，横轴为四个时间点，分组为1，2，3
def boxdata(df, fea):
    box1 = []
    box2 = []
    box3 = []
    for i,j in df[df["groupHPD"]==1].groupby("stay_id"):
        box1.append(j[fea].tolist())
    for i,j in df[df["groupHPD"]==2].groupby("stay_id"):
        box2.append(j[fea].tolist())
    for i,j in df[df["groupHPD"]==3].groupby("stay_id"):
        box3.append(j[fea].tolist())

    box1= pd.DataFrame(box1)
    box1 = box1.fillna(method="pad")
    box1 = box1.fillna(method="backfill").values
    box2 = pd.DataFrame(box2)
    box2 = box2.fillna(method="pad")
    box2 = box2.fillna(method="backfill").values
    box3 = pd.DataFrame(box3)
    box3 = box3.fillna(method="pad")
    box3 = box3.fillna(method="backfill").values
    return box1,box2,box3

def boxplot(fea,box1,box2,box3):
    x = [1, 6, 11, 16]
    x1 = [2, 7, 12, 17]
    x2 = [3, 8, 13, 18]
    x3 = [4, 9, 14, 19]
    
    box1_ = plt.boxplot(box1,positions=x,patch_artist=True,showmeans=True,showfliers=False,
                boxprops={"facecolor": "C0",
                          "edgecolor": "grey",
                          "linewidth": 0.5},
                medianprops={"color": "k", "linewidth": 0.5},
                meanprops={'marker':'+',
                           'markerfacecolor':'k',
                           'markeredgecolor':'k',
                           'markersize':5})
    plt_p(box1,0,1,x,"#0570b0")
    plt_p(box1,1,2,x,"#0570b0")
    plt_p(box1,2,3,x,"#0570b0")
    
    box2_ = plt.boxplot(box2,positions=x1,patch_artist=True,showmeans=True,showfliers=False,
                boxprops={"facecolor": "C1",
                          "edgecolor": "grey",
                          "linewidth": 0.5},
                medianprops={"color": "k", "linewidth": 0.5},
                meanprops={'marker':'+',
                           'markerfacecolor':'k',
                           'markeredgecolor':'k',
                           'markersize':5})
    plt_p(box2,0,1,x1,"#fc8d59")
    plt_p(box2,1,2,x1,"#fc8d59")
    plt_p(box2,2,3,x1,"#fc8d59")

    box3_ = plt.boxplot(box3,positions=x2,patch_artist=True,showmeans=True,showfliers=False,
                boxprops={"facecolor": "C2",
                          "edgecolor": "grey",
                          "linewidth": 0.5},
                medianprops={"color": "k", "linewidth": 0.5},
                meanprops={'marker':'+',
                           'markerfacecolor':'k',
                           'markeredgecolor':'k',
                           'markersize':5})
    plt_p(box3,0,1,x2,"#238b45")
    plt_p(box3,1,2,x2,"#238b45")
    plt_p(box3,2,3,x2,"#238b45")
    plt.xticks([1,2,3,4,6,7,8,9,11,12,13,14,16,17,18,19],["","","T1","","","","T2","","","","T3","","","","T4",""],fontsize=11)
    plt.xlim(0,20)
    plt.xlabel(fea,fontsize=11)
    plt.ylabel('value',fontsize=11)
    plt.grid(axis='y',ls='--',alpha=0.8)
    plt.legend(handles=[box1_['boxes'][0],box2_['boxes'][0],box3_['boxes'][0]],labels=['1','2',"3"])
    plt.tight_layout()
    plt.show()

def get_p(data,V1,V2):
    data = pd.DataFrame(data)
    p = stats.mannwhitneyu(data[V1],data[V2],alternative='two-sided').pvalue
    ymax = max(data.describe().T["max"])*0.3
    y_lst = [ ymax*i for i in [1,0.95,0.9]]
    return p,y_lst 

def plt_p(data,V1,V2,lst,color_):
    p,y_lst = get_p(data,V1,V2)
    y_ = (y_lst[V1]+y_lst[V1])/2
    if p <0.05:
        plt.plot([lst[V1],lst[V2]], [y_,y_], color=color_, linewidth=1)
        plt.text((lst[V1]+lst[V2])/2, y_, "*", color=color_,verticalalignment='center', horizontalalignment='center') #,**font_style)
    else:
        plt.plot([lst[V1],lst[V2]], [y_,y_], color=color_, linewidth=1)
```

# 发生速率 - 组间描述

```python
def divide_stage_lst(stage_lst):
    max_stage = max(stage_lst)
    s0 = stage_lst[0]
    lst = []
    lst_all = []
    num_lst_all = []
    num_lst = []
    num = 0
    for s in stage_lst:
        if s <= s0 and s0 in stage_lst[num:]:
            lst.append(s)
            num_lst.append(num)
            num +=1
            
        else:
            lst_all.append(lst)
            num_lst_all.append(num_lst)
            s0 = s
            lst = []
            num_lst = []
            num_lst.append(num)
            num +=1
            lst.append(s)
    lst_all.append(lst)    
    num_lst_all.append(num_lst)  
    return lst_all,num_lst_all
```

```python
from itertools import groupby
df_aki = pd.read_csv("/public/hanl/jupyter_dir/kidney_sepsis_penotype/00.data/disease_definition/sk_aki_onset_all.csv")
df_aki= df_aki.loc[:,[ 'stay_id', 'charttime', 'aki_stage']]
df_aki["charttime"] = pd.to_datetime(df_aki["charttime"], format='%d/%m/%Y %H:%M:%S')
df_aki = df_aki.sort_values(by=["stay_id","charttime"])
df_aki = df_aki[df_aki["stay_id"].isin(all_lst)]

df_AKI_duringTime = pd.DataFrame(columns=("stay_id","aki_stage","during_time","during_time_cum"))
for i,j in df_aki.groupby("stay_id"):
    stage_lst = j.aki_stage.tolist()
    t_lst = j.charttime.tolist()
    divide_stage, divide_t = divide_stage_lst(stage_lst)
    for k in range(len(divide_stage)):
        max_AKI_stage = max(divide_stage[k])
        during_time = (t_lst[divide_t[k][-1]] - t_lst[divide_t[k][0]])/np.timedelta64(1, 'h')
        during_time_cum = (t_lst[divide_t[k][-1]] - t_lst[0])/np.timedelta64(1, 'h')
        df_AKI_duringTime.loc[len(df_AKI_duringTime)] = [i, max_AKI_stage,during_time,during_time_cum]
df_AKI_duringTime = pd.merge(df_AKI_duringTime,df_type_filt,how ="inner",on="stay_id")

df_AKI_duringTime_add = pd.DataFrame()
for i,j in df_AKI_duringTime.groupby("stay_id"):
    j["during_time_cum_start"] = j["during_time_cum"].shift(+1)
    df_AKI_duringTime_add = pd.concat([df_AKI_duringTime_add, j],axis=0)
df_AKI_duringTime_add["during_time_cum_start"] = df_AKI_duringTime_add["during_time_cum_start"].fillna(0)
df_AKI_duringTime_add = df_AKI_duringTime_add.sort_values(["stay_id","during_time_cum_start","during_time_cum"])
df_AKI_duringTime_add["aki_stage"] = df_AKI_duringTime_add["aki_stage"].map(lambda x: int(x))
df_AKI_duringTime_add["stay_id"] = df_AKI_duringTime_add["stay_id"].map(lambda x: int(x))

df_AKI_duringTime_add.head(3)
```

```python
df_type = df_fea_base_add.loc[:,["stay_id","groupHPD"]]
df_fea_base_add = df_fea_base_add[p_lst]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_fea_base_std = scaler.fit_transform(df_fea_base_add) 
df_fea_base_std =  pd.DataFrame(df_fea_base_std)
print(df_fea_base_std.shape)
df_fea_base_std.columns = df_fea_base_add.columns.tolist()
df_fea_base_std = pd.concat([df_type, df_fea_base_std],axis=1)
print(df_fea_base_std.shape)

def data_get(data, stayid_lst):
    df = data[data["stay_id"].isin(stayid_lst)]
    df = df.groupby("groupHPD").agg(["mean","sem"])
    df.columns = ['_'.join(col) for col in df.columns] #.replace("median","mean")
    df = df.reset_index()
    df = df.drop(["stay_id_mean","stay_id_sem"],axis=1)

    from matplotlib import pyplot as plt 
    f_mean = [i for i in df.columns if i.endswith("_mean")]
    f_sem = [i for i in df.columns if i.endswith("_sem")]

    dfT = df[f_mean].T
    dfT.columns = [1,2,3]
    dfT = dfT.sort_values([3,2,1]).T

    import re
    f_mean = [i for i in dfT.columns if i.endswith("_mean")]
    f_sem = [re.sub("_mean$","_sem",i) for i in f_mean ]
    col_lst = [re.sub("_mean$","",i) for i in f_mean ]
    return df,f_mean, f_sem, col_lst
```
